```python
import os
from dotenv import load_dotenv
import pandas as pd
load_dotenv
path = os.getenv('DATASET_PATH')
```

```python
df =  pd.read_csv(path)
print(df.describe(include='all'))
df.head()
```

```python
df[df.isna()].sum()
```

```python
spilt = int(len(df)*0.8)
```

```python
train_set = df[:spilt]
test_set = df[spilt:]
train_set.to_csv('train_set.csv', index=False)
test_set.to_csv('test_set.csv', index=False)
```
